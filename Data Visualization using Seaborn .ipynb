```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(dir(sns))
```

```python
flights=sns.load_dataset("flights")
```

```python
flights
```

```python
rel=sns.relplot(x="month",y="passengers",hue="year",data=flights)
rel.set_xticklabels(rotation=45) 
#Plotted to figure out which month the number of passengers are high - also gives a relationship with the year. -
#it is observed that the No. of passengers are high during May to sept. 
```

```python
rel=sns.relplot(x="month",y="passengers",kind="line",hue="year",data=flights)
rel.set_xticklabels(rotation=45) 
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
sns.barplot(x="time", y="total_bill",hue="sex",data=tips)
#compares total bill during the time of the day - infers higher income during dinner & mostly male. 
```

```python
sns.lineplot(x="size",y="total_bill",data=tips,hue="sex")
```

```python
sns.boxenplot(x=tips["total_bill"])
```

```python
sns.boxenplot(x="day", y="total_bill", data=tips)
```

```python
sns.catplot(x="sex",y="total_bill",kind="strip",hue="time",data=tips)
```

```python
sns.catplot(x="sex",kind="count",hue="time",data=tips)
```

```python
catpl=sns.catplot(x="sex",y="total_bill",kind="point",hue="time",data=tips)
plt.title("relationship b/w sex & Total Bill at particular business hours")
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips)
```
